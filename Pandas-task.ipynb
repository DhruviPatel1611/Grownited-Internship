```python
import pandas as pd
```

```python
# 1️⃣ Create a Series from [10, 20, 30, 40].

s = pd.Series([10,20,30,40])
print(s)
```

```python
# 2️⃣ Create a Series with values [5, 6, 7] and index ["a","b","c"].

s = pd.Series([5,6,7], index = ['a','b','c'])
print(s)
```

```python
# 3️⃣ Create a Series from dictionary:

s = pd.Series({"Math": 85, "Sci": 90, "Eng": 88})
print(s)
```

```python
# 4️⃣ Create a Series with scalar value 100 repeated for 5 indices.

s = pd.Series(100, index=range(5))
print(s)
```

```python
# 5️⃣ Print first 3 values of a Series.

s = pd.Series([10,20,30,40,50,60,70])
print(s.head(3))
```

```python
# 6️⃣ Print last 2 values.

s = pd.Series([10,20,30,40,50,60,70])
print(s.tail(2))
```

```python
# 7️⃣ Get the datatype of a Series.

s = pd.Series([10,20,30,40,50,60,70])
print(s.dtype)
```

```python
# 8️⃣ Find number of elements in Series.

s = pd.Series([10,20,30,40,50,60,70])
print(s.size)
```

```python
# 9️⃣ Select value using label "b".

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
print(s.loc['b'])
```

```python
# 🔟 Select value using position 2.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
print(s.iloc[2])
```

```python
# 1️⃣1️⃣ Slice first 4 elements.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
print(s['a':'d'])
```

```python
# 1️⃣2️⃣ Filter values greater than 50.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
print(s[s>50])
```

```python
# 1️⃣3️⃣ Replace value 20 with 200.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
print(s.replace(20,200))
```

```python
# 1️⃣4️⃣ Rename index "a" to "alpha".

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
result = s.rename(index={'a':'alpha'})
print(result)
```

```python
# 1️⃣5️⃣ Sort a Series by values.

s = pd.Series([70,60,50,40,30,20,10], index=['a','b','c','d','e','f','g'])
print(s.sort_values())

s1 = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
print(s1.sort_values(ascending=False))
```

```python
# 🟡 MEDIUM LEVEL (16–30)

# 1️⃣6️⃣ Add 10 to every element in Series.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
print(s+10)
```

```python
# 1️⃣7️⃣ Multiply two Series together.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
s2 = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])

print(s*s2)
result = s.mul(s2)
print(result)
```

```python
# 1️⃣8️⃣ Find sum of Series.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
print(s.sum())
```

```python
# 1️⃣9️⃣ Find mean of values.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
print(s.mean())
```

```python
# 2️⃣0️⃣ Find median.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
print(s.median())
```

```python
# 2️⃣1️⃣ Find standard deviation.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
print(s.std())
```

```python
# 2️⃣2️⃣ Count non-null values.

s = pd.Series([10,None,30,None,50,None,70])
print(s.notnull())
print(s.notnull().sum())
```

```python
# 2️⃣3️⃣ Replace missing values with 0.

s = pd.Series([10,None,30,None,50,None,70])
s = s.fillna(0)
print(s)
print(s.isnull().sum())
```

```python
# 2️⃣4️⃣ Drop missing values.

s = pd.Series([10,None,30,None,50,None,70])
print(s)
s = s.dropna()
print(s)
```

```python
# 2️⃣5️⃣ Get top 3 largest values.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
result = s.sort_values(ascending=False)
print(result)
print(result.head(3))
```

```python
# 2️⃣6️⃣ Get 2 smallest values.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
result = s.sort_values(ascending=False)
print(result)
print(result.tail(2))
```

```python
# 2️⃣7️⃣ Apply lambda to square each value.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
result = s.map(lambda x : x * x)
result1 = s.apply(lambda x : x * x)
print(result)
print(result1)
```

```python
# 2️⃣8️⃣ Count frequency of unique values.

s = pd.Series([10,20,10,30,20,20,40,10])
print(s.value_counts())
```

```python
# 2️⃣9️⃣ Reindex Series to include new label "z".

s = pd.Series([10,20,30], index=['a','b','c'])
s['z'] = None
print(s)
```

```python
# 3️⃣0️⃣ Calculate cumulative sum of Series.

s = pd.Series([10,20,30,40,50,60,70], index=['a','b','c','d','e','f','g'])
print(s.cumsum())
```
