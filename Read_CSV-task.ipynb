```python
import pandas as pd
import numpy as np
```

```python
# 1️⃣ Load Dataset
# Load Movieratingdata1.csv and show all rows & columns.

df = pd.read_csv("Movieratingdata1.csv")
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)

print(df)
```

```python
# 2️⃣ Custom Missing Values

# While loading, treat
# "not available" and "not mension" as NaN in PRODUCTION_YEAR.

df.replace(["not available","not mension"],np.nan,inplace=True)
print(df)
```

```python
# 3️⃣ Replace Text Missing Values

# Replace "not available", "not mension", "not value" in whole dataframe with np.nan.

df.replace(["not available","not mension","not value"],np.nan,inplace=True)
print(df)
```

```python
# 4️⃣ Count Total Missing Values

# Find total number of missing values in the dataset.

print(df.isnull().sum().sum())
```

```python
# 5️⃣ Column-wise Missing Count

# Show missing values count for each column.

print(df.isnull().sum())
```

```python
# 6️⃣ Drop Rows Where All Values Missing

# Remove rows where every column is NaN.

df.dropna(inplace=True,how="all")
print(df)
```

```python
# 7️⃣ Drop Columns with Any Missing Value

# Remove columns that contain at least one NaN.

df.dropna(axis=1,inplace=True,how="any")
print(df)
```

```python
# 8️⃣ Fill Missing Movie ID

# Fill missing MOVIE_ID with 0.

df.replace("not value",np.nan,inplace=True)
df.fillna({"MOVIE_ID":0},inplace=True)
print(df)
```

```python
# 9️⃣ Fill Missing Production Year

# Replace missing PRODUCTION_YEAR with 2000.

df.replace(["not available","not mension"],np.nan,inplace=True)
df.fillna({"PRODUCTION_YEAR":2000},inplace=True)
print(df)
```

```python
# 🔟 Fill Missing Ratings with Mean

# Convert RATING to float and fill missing values using average rating.

df["RATING"] = df["RATING"].astype(float)
df["RATING"] = df["RATING"].fillna(df["RATING"].mean())
print(df)
```

```python
# 1️⃣1️⃣ Replace Wrong Entry

# Replace "1390411" with "The Batman".

df.replace('1390411','The Batman',inplace=True)
print(df)
```

```python
# 1️⃣2️⃣ Rename Columns

# Rename columns to lowercase.

df.rename(columns=str.lower,inplace=True)
print(df)
```

```python
# 1️⃣3️⃣ Change Data Type

# Convert PRODUCTION_YEAR to integer.

# print(df)
df["production_year"] = df["production_year"].astype(int)
print(df)
```

```python
# 1️⃣4️⃣ Remove Duplicate Rows

# Delete duplicate rows from dataset.

# df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
print(df)
```

```python
# 1️⃣5️⃣ Show Data Types

# Display datatype of each column.

print(df.dtypes)
```

```python
# 1️⃣6️⃣ Reset Index

# Reset the dataframe index after cleaning.

df = pd.read_csv("Movieratingdata1.csv")
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
df.reset_index(drop=True,inplace=True)
print(df)
```

```python
# 1️⃣7️⃣ Drop Movie Title Column

# Remove MOVIE_TITLE column.

df = pd.read_csv("Movieratingdata1.csv")
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
df.drop("MOVIE_TITLE",axis=1,inplace=True)
print(df)
```

```python
# 1️⃣8️⃣ Fill All Remaining NaNs

# Fill all remaining NaN values with "Unknown".

df.fillna("Unknown",inplace=True)
print(df)
```

```python
# 1️⃣9️⃣ Describe Dataset

# Show summary statistics of numeric columns.

print(df.describe())
```
